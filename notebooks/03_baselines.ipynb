```python
from src.data_utils import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_data(regions='all', years='all')
```

```python
regions = data['region'].unique()
names = data.columns
variables = ['msl_mean', 'msl_p15', 'msl_p85', 'msl_std',
       't2m_mean', 't2m_p15', 't2m_p85', 't2m_std', 'tp_mean', 'tp_p15',
       'tp_p85', 'tp_std', 'u10_mean', 'u10_p15', 'u10_p85', 'u10_std',
       'v10_mean', 'v10_p15', 'v10_p85', 'v10_std', 'wind_speed_mean',
       'wind_speed_p15', 'wind_speed_p85', 'wind_speed_std', 'extent_mkm2']
climatology_tables = {}
```

```python
climatology_extent = data.groupby(['region', data['date'].dt.dayofyear])['extent_mkm2'].mean().unstack().T
```

```python
for region in regions:
    for variable in variables:
        region_data = data[data['region'] == region]
        climatology_table = region_data.groupby(region_data['date'].dt.dayofyear)[variable].mean().T
        climatology_tables[f'{region}_{variable}'] = climatology_table
```
